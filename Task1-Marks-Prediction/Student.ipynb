```python
#Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Importing data set
data = pd.read_csv('Student_Data.csv')
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
```

```python
data.shape
```

```python
X
```

```python
y
```

```python
#Splitting the data set into training and test data set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
X_train
```

```python
X_test.shape
```

```python
#Training the Linear regression model on training set
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
#Predicting the test set results
pred = lr.predict(X_test)
```

```python
#Visualizing the training set results
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, lr.predict(X_train), color='blue')
plt.title('Marks VS Preparation(training set)')
plt.xlabel('Preparation(in hours)')
plt.ylabel('Marks')
```

```python
'''Visualizing the test set results'''
plt.scatter(X_test, y_test, color='red')
plt.plot(X_train, lr.predict(X_train), color='blue')
plt.title('Marks VS Preparation(Testing set)')
plt.xlabel('Preparation(in hours)')
plt.ylabel('Marks')
```

```python
'''Comparing actual data with predicted data
 As you can see te model has decently well.'''
df = pd.DataFrame({'Actual ': y_test, 'Predicted ':pred})
df
```

```python
'''What will be the marks of students studying for 9.25 hours? '''
lr.predict([[9.25]])
'''So student studying for 9.25 hours can score 93.8927 marks in the exaam '''
```

```python
'''Evaluating the model
The final step involves evaluating the model here we have used mean squared error'''
from sklearn import metrics
from sklearn.metrics import accuracy_score
metrics.mean_absolute_error(y_test, pred)
```
