```python
'''FROM tHE above analysis it is clear that the data is of united states '''
data = data.drop(['Country', 'Postal Code'], axis=1)
data.head()
```

```python
#SOME visualisations for better insights
sns.jointplot(x='Sales', y='Profit', data=data)
plt.show()
''' FROM the plot it can be seen that if sales are less it directly affects the profit
```

```python
sns.jointplot(x='Quantity', y='Profit', data=data, color='green')
plt.show()
```

```python
sns.jointplot(x='Discount', y='Profit', data=data, color='magenta')
plt.show()
```

```python
fig_dms=(9,5)
fig, ax = plt.subplots(figsize=fig_dms)
sns.barplot(x='Region', y='Profit', data=data, ax=ax)
plt.show()
'''On a larger picture company should focus on central region as profits seems less in this region as compared to other regions'''
```

```python
p = data.groupby(by='Segment')[['Profit']].sum()
r = p.sort_values(by=['Profit'], ascending=True)
r.plot(kind='bar', figsize=(9,5))
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='Region', y='Profit', hue='Segment', data=data, palette='coolwarm')
plt.show()
'''In central region Consumer segment is very weak, whereas in south region Home office is no doing well.
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Region', y='Sales', hue='Category', data=data, palette='rainbow')
plt.show()
'''Sales of Office supplies are not good in all the regions, technology is doing well in all the regions '''
```

```python
subcats = data.groupby(by='Sub-Category')[['Sales', 'Profit']].sum()
subcats.plot(kind='bar', figsize=(10,6), color=['green', 'maroon'])
plt.show()
'''Copiers has the highest profits whereas tables are showing negative profits, sales of chaiirs and phones are highest and lowest in case of fasteneres '''
```

```python
#State wise profits
plt.figure(figsize=(20,12))
data.groupby(by='State')['Profit'].sum().plot(kind='bar', color='lightskyblue')
plt.show()
'''Texas has lowest profit and california shows highest profits '''
```
